#Python Basics

1.What is Python, and why is it popular?
 - Python is a high-level, interpreted programming language known for its simple and readable syntax. It is popular because it is easy to learn, has a large collection of libraries, supports multiple programming areas (such as web development, data science, and AI), and has strong community support. Its versatility and readability make it suitable for both beginners and professionals.
  
2.What is an interpreter in Python?
 - An interpreter in Python is a program that executes Python code line by line by translating it into machine-readable instructions so the computer can run it.

3.What are pre-defined keywords in Python?
 - Pre-defined keywords in Python are reserved words that have special meanings and purposes in the language. They are used to define the structure and rules of Python programs, and cannot be used as variable names or identifiers. Examples include if, while, for, class, and return.

4.Can keywords be used as variable names?
 - No, keywords cannot be used as variable names in Python because they are reserved words that have special meanings and are used to define the syntax and structure of the language.

5.What is mutability in Python?
 - Mutability in Python refers to the ability of an object to be changed after it is created. If an object is mutable, its contents (like elements or values) can be modified without creating a new object. Examples of mutable objects include lists, dictionaries, and sets.

6.Why are lists mutable, but tuples are immutable?
 - Lists are mutable because Python stores them in a flexible memory structure that allows you to change their elements, add new items, or remove items after the list is created.Tuples are immutable because Python stores them in a fixed memory structure. This means that once a tuple is created, its size and elements cannot be changed. You cannot add, remove, or modify items in a tuple.

7.What is the difference between “==” and “is” operators in Python?
 - "==" operator: Checks whether the values of two objects are equal. It tells you if the data stored in the objects is the same.
 "is" operator: Checks whether two variables refer to the same object in memory. It tells you if both variables point to the exact same object, not just the same value.

8.What are logical operators in Python?
 - Logical operators are used to combine multiple conditions or boolean expressions and return True or False based on the logic.

The main logical operators in Python are:

 - "and"→ Returns True if both conditions are True.
 - "or" → Returns True if at least one condition is True.
 - "not" → Reverses the result; returns True if the condition is False, and     vice versa.

9.What is type casting in Python?
 - Type casting in Python is the process of converting a value from one data type to another. It allows you to change the type of a variable so it can be used in different operations.

Example types of type casting:

 - Converting an integer to a float
 - Converting a string to an integer
 - Converting a list to a tuple

10.What is the difference between implicit and explicit type casting?
 - implicit casting happens automatically when python converts one type to another for example, converting an (int) to a (float) during arithmetic.
 - explicit casting is done manually by the programmar using function like int, float, string.
 implicit is automatic and safe, while explicit gives more control but must be use carefully.

11.What is the purpose of conditional statements in Python?
 - Conditional statements in Python are used to make decisions in a program. They allow the program to execute certain blocks of code only if a specific condition is true, and optionally execute other blocks if the condition is false. This helps in controlling the flow of the program based on different situations or inputs.

12.How does the elif statement work?
 - elif allows checking multiple conditions one by one after an initial if, and only the first true condition is executed.

13.What is the difference between for and while loops?
 - A for loop is used when the number of iterations is known in advance, such as looping over a sequence like a list, tuple, or range. It executes a block of code for each item in the sequence. A while loop, on the other hand, is used when the number of iterations is not known beforehand and the loop continues as long as a specified condition is true.

14.Describe a scenario where a while loop is more suitable than a for loop.
 - A while loop is more suitable than a for loop when the number of iterations is not known in advance and depends on a condition being met. For example, consider a program that asks a user to enter a password. The program should keep asking until the user enters the correct password. Since we don’t know how many attempts the user will take, a while loop is ideal. It continues to execute as long as the entered password is incorrect and stops automatically when the correct password is entered.

5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

```python
my_list = [1, 2, 3]
print("Original list:", my_list)
my_list[1] = 20
print("Modified list:", my_list)
my_tuple = (1, 2, 3)
print("Original tuple:", my_tuple)

# Attempting to change an element in the tuple
try:
    my_tuple[1] = 20
except TypeError as e:
    print("Error:", e)
```

6. Write a function to demonstrate the behavior of mutable and immutable arguments.

```python
# Function to modify a mutable object (list)
def modify_list(lst):
    lst.append(100)
    print("Inside function (list):", lst)

# Function to modify an immutable object (integer)
def modify_number(num):
    num += 100
    print("Inside function (number):", num)

# Mutable object example
my_list = [1, 2, 3]
print("Before function call (list):", my_list)
modify_list(my_list)
print("After function call (list):", my_list)
print()

# Immutable object example
my_number = 50
print("Before function call (number):", my_number)
modify_number(my_number)
print("After function call (number):", my_number)
```

13.. Write a Python program to find the sum of all even numbers between 1 and 50.

```python
sum_even = 0
for i in range(1, 51):
    if i % 2 == 0:
        sum_even += i
print("Sum of even numbers from 1 to 50 is:", sum_even)
```

14.Write a program to reverse a string using a while loop.

```python
text = input("Enter a string: ")
i = len(text) - 1
reversed_text = ""
while i >= 0:
    reversed_text += text[i]
    i -= 1
print("Reversed string:", reversed_text)
```

15.Write a Python program to calculate the factorial of a number provided by the user using a while loop

```python
num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial *= i
    i += 1

print(f"Factorial of {num} is {factorial}")
```
